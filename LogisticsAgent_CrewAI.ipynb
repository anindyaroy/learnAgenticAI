Goal of the project is to analyse logisitcs data and develop optimisation strategies for a specific logistics industry problem. For example optimising delivery routes or inventory management.

```python
from crewai import Crew, Task, Agent, LLM
from crewai_tools import SerperDevTool

# Define the Agents
logistics_analyst = Agent(
    role = "Logistics Analyst",
    goal = "Research the current state of logistics operations in the Logistics industry",
    backstory = "Experienced Logistics analyst who has executed many projects in the logistics industry researching the current state of the market",
    tools = [SerperDevTool()],
    verbose=True,
)

strategist_backstory = ""
optimisation_strategist = Agent(
    role = "Optimization Strategist",
    goal = "Create an effective strategy to optimise Logistics operations using market research inputs",
    backstory= f"Expert in optimising Logistics operations {strategist_backstory}",
    verbose = True,
)
```

```python
research_task = Task(
    description="Research the state of the operations in logistics industry focusing on the route efficiency and inventory turnover trends",
    expected_output= """
    Create a research report of the current best practices in logistics operations for best route efficiency and inventory turnover needs.
    Provide output in an Agent consummable format
    """,
    agent=logistics_analyst
)

optimisation_task = Task(
    description= "Create an optimisation strategy based on instghts provided as input",
    expected_output= "Create an effective strategy that will enable the optimisation of the logictics operations and maximise profits. Provide output in Markdown",
    agent=optimisation_strategist
)
```

```python
crew = Crew (
    agents=[logistics_analyst,optimisation_strategist],
    tasks=[research_task,optimisation_task],
    planning=True
)

crew.kickoff()
```
